```python
 age = [13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 33, 35, 35, 35, 35, 36, 40, 45, 46, 52, 70]
```

1. What is the mean value of the data?

```python
sum = 0
for element in age:
  sum += element

mean = sum / len(age)
print(round(mean, 2))
```

2. What is the median value of the data?

```python
#age.sort()
#median: Since there are 27 values in the dataset, the median will be the middle value when the values are arranged in ascending order.
# If there are an odd number of values, the median will be the middle value.
# If there are an even number of values, the median will be the average of the two middle values.
numberOfValues = len(age)
middleElementsList =[]
sum = 0
counter = 0

if numberOfValues % 2 == 0 :
    middleElementsList.append(age[int((numberOfValues - 1)/2)])
    middleElementsList.append(age[int(((numberOfValues - 1)/2) + 1)])
else:
    middleElementsList.append(age[int((numberOfValues - 1)/2)])

for element in middleElementsList:
    sum += element
    counter += 1

medium = sum / counter
print(medium)
```

3. What is the mode of the data? Comment on the data’s modality (i.e., bimodal, trimodal, etc.).

```python
#This is bimodal because we have two values that repeat itself often
countOccurences = {}
modes = []

for element in age:
  if element not in countOccurences:
      countOccurences[element] = 1
  else:
    countOccurences[element] += 1

maxValue = max(countOccurences.values())

for key in countOccurences.keys():
    if countOccurences[key] == maxValue:
        modes.append(key)

print(modes)
```

5. Can you find the first quartile (Q1) and the third quartile (Q3) of the data?

  1. First Quartile (Q1 - 25th percentile): The value below which 25% of the data
    falls.
  2. Second Quartile (Q2 - 50th percentile, Median): The middle value of the dataset (also the median).
  3. Third Quartile (Q3 - 75th percentile): The value below which 75% of the data falls.
These values help measure data spread and identify outliers when used in box plots.

```python
numberOfValues = len(age)
q1=(0.25*(numberOfValues+1))
q3=(0.75*(numberOfValues+1))

print("Position of Q1", q1)
print("Position of Q3", q3)

print("Q1 value", age[6])
print("Q3 value",age[20])

print("IQR", age[20] - age[6])

#25% of people are 20 or younger
#75% of people are 35 or younger
#50% of peoples age is spread across 15 years
```

```python
#Second way
import numpy as np

#calculating quartile is done by 0.25 * len(data); and averaging the values between those positions

Q1 = np.percentile(age,  25, method='nearest')
q2 = np.percentile(age, 50, method='nearest')
Q3 = np.percentile(age, 75, method='nearest')
print("Quarter 1:", Q1)
print("Quarter 3:", Q3)

#   When you use method="nearest" in np.percentile(), it does not round up or down mathematically.
#   Instead, it picks the closest actual value from the dataset.
#   Anotther example:
#   position 7.25 is between index 6 (20) and index 7 (21).
#   Using method="nearest":
#   20 (index 6) is closer to 7.25 than 21 (index 7).
#   So Q1 = 20.
```

1. Calculate the mean, median, and standard deviation of age and %fat

Standard Deviation (σ or SD) measures how much the values in a dataset deviate

from the mean (average). It shows how "spread out" the data is.

A low SD means the data points are closer to the mean.

A high SD means the data points are more spread out from the mean.

```python
age_mean = round(df["age"].mean(), 2)
fat_mean = round(df["fat"].mean(), 2)
print("Mean for age:", age_mean)
print("Mean for fat:", fat_mean)

medianAge = df["age"].median()
medianFat = df["fat"].median()
print("Median for age:", medianAge)
print("Median for fat:", medianFat)

stdAge = round(df["age"].std(), 2)
stdFat = round(df["fat"].std(), 2)
print("SD for age:", stdAge)
print("SD for fat:", stdFat)
```

5. Draw a quantile plot
You can pass different values for the line argument:

1. '45'	45-degree line	A straight diagonal line from (0,0) to (max, max). Used for comparing data against a normal distribution.
2. 's'	Standardized Line	Uses the sample mean and standard deviation to adjust the Q-Q plot. Helps compare deviations from normality.
3. 'r'	Regression Line	Fits a least-squares regression line through the points. Useful for seeing the overall trend.
4. 'q'	Quartile Line	Fits a line passing through the first and third quartiles (Q1 & Q3). Helps compare the center distribution.

1. line='45': Best for normal distribution comparisons.
2. line='s': If your data has a different mean/variance but should follow a normal shape.
3. line='r': To check the trend of data points.
4. line='q': When you want to focus on quartiles rather than extremes.

1. A Q-Q plot (Quantile-Quantile plot) is used to compare the distribution of a dataset against a theoretical distribution (usually a normal distribution).

2. In this case, the Q-Q plot is comparing the "Fat Percentage" data against a normal distribution.

3. X-axis (Theoretical Quantiles) → Represents what the data should look like if it follows a normal distribution.

4. Y-axis (Sample Quantiles) → Represents the actual data values from the dataset.

5. Red Line (line='q') → A line passing through Q1 and Q3 (first and third quartiles). It helps compare the center distribution of the data.

6. Blue Dots → Each dot represents a data point plotted against its theoretical quantile.

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

def qq_plot(data, title):
    sm.qqplot(data, line='q')
    plt.title(f'Q-Q Plot of {title}')
    plt.xlabel("Theoretical Quantiles")
    plt.ylabel("Sample Quantiles")
    plt.grid()
    plt.show()

# Q-Q plot for 'age'
#qq_plot(df['age'], 'Age')

# Q-Q plot for 'fat_percentage'
qq_plot(df['fat'], 'Fat Percentage')
```

CONCLUSION:
1. Most blue dots are closely following the red line in the middle → The fat percentage data appears to be somewhat normally distributed in the center.
2. Some dots at the upper end deviate from the line → There might be some higher values (outliers) in fat percentage.
3. The lower part of the plot is well aligned with the red line, suggesting the lower values of fat percentage fit a normal distribution better than the upper values.
