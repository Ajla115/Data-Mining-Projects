```python
from sklearn.preprocessing import minmax_scale
import seaborn as sns
import matplotlib.pyplot as plt

usd_goal=file3["usd_goal_real"]

scaled_data=minmax_scale(usd_goal, axis=0) # scale each column independently

fig,ax=plt.subplots(1,2)
#Creates a figure with 1 row and 2 columns of plots (side by side).
#ax is an array with two axes (ax[0] and ax[1]), one for each plot.
sns.kdeplot(usd_goal, ax=ax[0])
ax[0].set_title("Original Data")
sns.kdeplot(scaled_data, ax=ax[1])
ax[1].set_title("Scaled data")
```

```python
file3["goal"] = pd.to_numeric(file3["goal"], errors="coerce")
file3 = file3.dropna(subset=["goal"])

file3 = file3.copy()
file3["goal"] = pd.to_numeric(file3["goal"], errors="coerce")
file3.dropna(subset=["goal"], inplace=True)
file3["goal"].fillna(file3["goal"].mean(), inplace=True)


#msm da mi ovo nije potrebno vise, ali neka stoji sad za sad pa mogu obrisati
```

```python
goal=file3["goal"]

scaled_data_2=minmax_scale(goal, axis=0)

fig, ax=plt.subplots(1,2)
sns.kdeplot(goal, ax=ax[0])
ax[0].set_title("Original Data")
sns.kdeplot(scaled_data_2, ax=ax[1])
ax[1].set_title("Scaled data")
```

```python
from scipy import stats

file3["usd_pledged_real"] = pd.to_numeric(file3["usd_pledged_real"], errors="coerce")

#Creates a Boolean mask (True or False) for the "usd_pledged_real" column.
#Only keeps values where usd_pledged_real is greater than 0 (ignores projects with zero pledges).
index_of_positive_pledges = file3["usd_pledged_real"] > 0

# Uses .loc[] to filter only the rows where index_of_positive_pledges is True.
# Extracts a new Series (positive_pledges) that contains only positive values.
positive_pledges = file3["usd_pledged_real"].loc[index_of_positive_pledges]

# Applies Box-Cox transformation to normalize the distribution of positive pledge values.
# Box-Cox is useful when data is skewed (not normally distributed).
# The [0] at the end extracts the transformed values
normalized_pledges = stats.boxcox(positive_pledges)[0]

fig, ax=plt.subplots(1,2)
sns.kdeplot(positive_pledges, ax=ax[0])
ax[0].set_title("Original Data")
sns.kdeplot(normalized_pledges, ax=ax[1])
ax[1].set_title("Normalized data")
```

```python
file3["pledged"] = pd.to_numeric(file3["pledged"], errors="coerce")
index_of_pledges = file3["pledged"] > 0

positive_pledges = file3["pledged"].loc[index_of_pledges]

normalized_pledges = stats.boxcox(positive_pledges)[0]

fig, ax=plt.subplots(1,2)
sns.kdeplot(positive_pledges, ax=ax[0])
ax[0].set_title("Original Data")
sns.kdeplot(normalized_pledges, ax=ax[1])
ax[1].set_title("Normalized data")
```
